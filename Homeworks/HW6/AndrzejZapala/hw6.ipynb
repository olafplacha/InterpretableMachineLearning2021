Korzystałem z datasetu titatnic.
Wytrenowałem 4 modele szacujące prawdopodobieństwa przeżycia katastrofy.

**Trening 4 modeli**:
* Lightgbm (wybrany model drzewiasty)
* Sieć neuronowa
* Regresja Logistyczna
* Las losowy

**Powyższe modele osiągnęły zadawalające wyniki w metryce roc auc**:
* zbior treningowy Lightgbm: 0.95
* zbior testowy Lightgbm: 0.90
* zbior treningowy siec neuronowa: 0.86
* zbior testowy siec neuronowa: 0.87
* zbior treningowy regresja logistyczna: 0.83
* zbiór testowy regresja logistyczna: 0.85
* zbior treningowy las losowy: 0.87
* zbior testowy las losowy: 0.88

### Statistical parity ratio
Dla wszystkich modeli metryka statistical parity ratio była zdecodwanie niższa dla mężczyzn, było to spowodowane tym, że kobiety charakteryzowały się znacznie większą przeżywalnością. Stąd wytrenowane modele "dyskryminują" mężczyzn dająć im niższe szanse na przeżycie.

## Accuracy equality ratio
Wszystkie 4 modele charakteryzują się podobną dokładnościa w grupie mężczyzn i w grupie kobiet. Zauważamy, że model Lightgbm ma najwyższe wartości tej metryki, charakteryzował się on najwyższą metryką roc auc. Ranking uśrednionej metryki accuracy equality ratio dla kobiet i mężczyzn dla każdego modelu opdowiada rankingowi metryki roc auc. Stąd modele nie dyskriminują pod względem tej metryki. Jednakże dla każdego modelu oprócz regresji logistycznej wartości dla mężczyzn były mniejsze. Modele otrzymywał lepsze rezultaty dla kobiet (oprocz regresji logistycznej), pomimo że mężczyzn było więcej w datasecie. 

## Predictive equality ratio
Wszsystkie modele charakteruzyją się wyższą metryką predictive equality ratio dla kobiet. Pokazuje to, że dla każdego modelu FPR jest wyższy dla kobiet. Stąd pomimo sporej dysproporcji w przeżywalności kobiet i mężczyzn w datasecie, modele jeszcze bardziej faworyzują kobiety pod względem przeżywalności. Częściej przeszacowują przeżywalność kobiet niż mężczyzn. Widzimy, że predictive equality ratio jest odwrotnie proporcjonalna do ogólnej skuteczności modelu np. do metryki roc auc. 

## Equal opportunity ratio
Każdy model dyskryminował mężczyzn pod względem metryki predictive equality ratio. Metryka przedstawia TPR. Nasuwają się podobne wnioski jak w poprzednim akapicie. Model przerzacowuje przeżywalność kobiet, niedoszacowując mężczyzn. Najskuteczniejszy model Lightgbm charakteryzuje się najmniejszą dysproporcją w wartości tej metryki dla mężczyzn i kobiet. 

## Podsumowanie
Modele dyskryminowały mężczyzn pod względem wszsystkich metryk oprócz metryki accuracy equality ratio. Sporą różnicę wartość statistical parity ratio można tłumaczyć różną średnią przeżywalnością kobiet i mężczyzn w danych, co może nie być błędęm modelu. Jednakże w ogólności modele przeszacowywały przeżywalność kobiet niedoszacowując przeżywalność mężczyzn. Najskuteczniejszy model charakteryzował się najmniejszymi różnicami w metyrkach fairness.
