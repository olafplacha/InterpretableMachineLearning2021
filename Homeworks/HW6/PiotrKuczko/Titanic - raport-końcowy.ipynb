# Titanic

Dane zawierają takie dane jak:
 - czy pasażer przeżył,
 - klasą która podróżował pasażer,
 - płeć,
 - wiek,
 - liczba rodzeństwa/małżonków na pokładzie,
 - liczba rodziców/dzieci na pokładzie,
 - cena biletu.

Naszym celem jest przewidzenie czy pasażer przeżył katastrofę na tytaniku na podstawie różnych obserwacji. Można się domyślać, że istotna mogła być klasą którą podróżował pasażer, płeć oraz wiek pasażera. Do mierzenia uczciwości modelu, wiek zamieniam to wartość 0 lub 1. 0 dla osób poniżej 18 roku życia, 1 dla pozostałych osób. Mierząc Pclass jako klasę uprzywiliowaną ustawiam klasę 1, natomiast jako klasy nie uprzywiljowane klasy 2 i 3. Na początku policzmy sobie statistical_parity_difference() za pomocą AIF360.

Teraz popatrzmy na 'statistical_parity_ratio'

Co ciekawe statistical_parity_difference() różni się w zależności od modelu, jednak widzimy, że dane są najbardziej nie uczciwe pod względem płci, znacznie mniej jeśli chodzi o wiek. Natomiast nie występowała stronniczość jeśli chodzi o klasę w jakiej podróżował pasażer.

Zobaczmy teraz metrykę 'equal opportunity ratio' na tych danych. Modele XGBoost nie są stronnicze dla wieku, natomiast RandomForest jest. Możemy tak sądzić ponieważ dane przekraczają wartość 0.8 w tych miejscach. Widać, że RandomForest jest mocno nie tolerancyjnym modelem.

Poniżej jest tabelka przedstawiająca 'predictive equality ratio'. Prawie wszystkie wartości są poniżej 0.8 (oprócz wieku w modelu XGBoost z głębokością 2). Co może świadczyć o tym, że ten model prawie w ogóle nie bierze wieku pod uwagę przy predykcji.

W tym przypadku zarówno dane jak i modele, które nauczyły się tych danych są mocno nieuczciwe. Choć w sumie ciekawe jest to, że dane nie były stronnicze pod względem Pclass, natomiast modele już stały się dużo bardziej nieuczciwe pod tym względem.
